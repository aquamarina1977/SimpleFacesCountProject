Проект из области компьютерного зрения. Этот код определяет количество лиц на фотографии и является вариацией многочисленных проектов в интернете для определения человеческих лиц или количества лиц. Файл haarcascade_frontalface_default.xml взят из репозитория на GitHub, размещённом по адресу: https://github.com/opencv/opencv.git

```python
import cv2
import numpy as np
import pandas as pd
from google.colab.patches import cv2_imshow
from PIL import Image
```

```python
import zipfile

def unzip(archive):
    zfile = zipfile.ZipFile(archive, 'r')
    for i_file_name in zfile.namelist():
        zfile.extract(i_file_name)
    zfile.close()
```

```python
unzip('/content/opencv-4.x.zip')
```

```python
casc_path = '/content/opencv-4.x/data/haarcascades/haarcascade_frontalface_default.xml'
```

```python
face_cascade = cv2.CascadeClassifier(casc_path)
```

```python
input_ = Image.open('/content/image1.jpg').convert('L')
image = np.array(input_, 'uint8')
faces = face_cascade.detectMultiScale(image, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))
```

```python
print("Найдено {0} лиц(-а).".format(len(faces)))

for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)

cv2_imshow(image)
cv2.waitKey(0)
```
